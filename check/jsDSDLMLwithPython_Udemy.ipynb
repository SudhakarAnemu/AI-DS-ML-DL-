## 21 Activity - Covariance and Correlation 

1 Measuring covariance

-> Think of the datasets for the two variables as high-dimensional vectors
-> convert these two vectors of variances form the mean
-> take the dot product(cosine of the angel between them) of the two vectors
-> divide by the sample size

2 Interpreting covariance is hard

-> We know a small covariance, close to 0, means there is not much correlation between the two variables
-> and large covarinaces - that is, far from 0 (could be negative for inverse relationships) mean there is a correlation. 
-> but how large is "large" ?

3 That's where correlation comes in !
-> Just divide the covariance by the standard deviations of both variables, and that normalizes things.
-> so a correlation of -1 means a perfect inverse correlation 
-> Correlation of 0: no correlation 
-> Correlation 1 : perfect correlation 

4 Remember - correlation does not imply causation!
-> only a controlled, randomized experiment can give you insights on causation.
-> Use correlation to decide what experiments to conduct!


```
construct a def de_mean(x):
  xmean = mean(x)
  return [xi - xmean for xi in x]

def covariance(x,y):
  n = len(x)
  return dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covarinace(pageSpeeds, purchaseAmount)
```

## 22 Conditional Probability (Exercise)
ex : Amazon - ppl order this also order this, ppl viewed this also viewed this.
P(B/A) = P(A,B)/P(A)

1 if I have two events that depend on each other, what's the probability that both will occur?
Notation - P(A,B) is the probability of A and B both occuring
P(B/A) : Probability of B given that A has occured
We know : 
P(B/A) = P(A,B)/P(A)

ex : A = passing the first test : 60%, B = Passing the second test : 80%
Q : What is the percentage of students who passed the first test aslo passed the second?
-> P(B/A) = P(A/B)/P(A) = 0.6/0.8 = 0.75
75% of the students who passed the first test passed the second one.



```
from numpy import random
random.seed(0) # will get consistent result every time we run the code

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
print("Totals : ", type(totals))
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
print("Totals : ", type(purchases))
totalPurchases = 0
for _ in range(100000): # _ is a placehodler it does not care .. it can be anything
  ageDecade = random.choice([20,30,40,50,60,70])
  purchaseProbability = float(ageDecade) / 100.0
  totals[ageDecade] += 1
  if (random.random() < purchaseProbability):
    totalPurchases += 1
    purchases[ageDecade] += 1
 
print("Totals : ",totals)
print("Purchases : ", purchases)
print("Total purchases : ", totalPurchases)
```

. Fit a line t oa data set of observations

. I don't know why they call it "regression". it's really misleading. You can use it to predict points in the future, the past, whatever. in fact time usually has nothing to do with it. 

How does it works?
. Usually using "least squares"
. Minimizes the squared-error between each point and the line
. y = mx+b : slope - intercept equeation 
. slope is the correlation between the two variables times the standard deviation in Y. all divided by the standard deviation in X. 
. intercept is the mean of Y minus the slope times the mean of x

. least squares minimizes the sum of squared errors. 
. the is the same as maximize the likelihood of the observed data if you start thinking of the problem in terms of probabilites and probability distribution functions
. this is sometimes called "maximum likelihood estimation"


. Gradient descent is an alternative method to least squares
. basically iterates to find the like that best follows the contours defined by the data. 
. can make sense when dealing with 3d data
. Least squares is a perfectly good choice. 

--- Measuring error with r-squared.
. How do we measure how well our line fits our data?
. R-squared (aka coefficient of determination) measures:

The fraction of the total varition in Y that is captured by the model

Computing r-squared 
1.0 - sum of squared errors/sum of squared variation from mean

## 33 K-Means clustering - Un supervised
<br>
Attemps to split data into K groups that are closest to K centroids
<br>Un supervised learning - uses only the positions of each data point
<br>can uncover interesting groupings of ppl /things / behavior
<br> ex : where do millionaries live? what genres of music/moves/etc naturally fall out of data?. create your wn sterotypes from demographic data
<br>How it works
<br>1. Randomly pick K centroids(k-means) 
<br>2. Assign each data point to the centroid it's closest to
<br>3. recompute the centroids based on the avg positio of each centroid's points
<br>4. iterate until points stop chaning assignment to centroids. 
<br>If you want to predict the cluster for new points, just find the centroid they're closest to.

<br> K-Means clustering gotchas
<br>Choosing K : try increasing K values until you stop getting large reductions in squared error(distances from each point to their centroids)
<br>Avoiding local minima : 
<br>A. the random choice of initial centroids can yield diff results
<br>B. run it a few times just to make sure your initial results are not wacky
<br> Labeling the clusters : 
<br> a. K-Means does not attempt to assign any meaning to the clusters you find
<br> it is up to you dig into the data and try to determine that. 




## 34 Activity - Clustering people base on income and age

```
# create fake income/age clusters for N people in k clusters
def createClusterData(N, k):
  np.random.seed(10)
  pointsPerCluster = float(N)/k
  X = []
  for i in range(k):
    incomeCentroid = np.random.uniform(20000.0, 200000.0)
    ageCentroid = np.random.uniform(20.0, 70.0)
    for j in range(int(pointsPerCluster)):
      X.append([np.random.normal(incomeCentroid, 10000.0), np.random.normal(ageCentroid, 2.0)])
  X = np.array(X)
  return X
```

```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

data = createClusterData(100,5)

```

```
data
```

```
scale(data)
```

```
model = KMeans(n_clusters=5)
model = model.fit(scale(data))
# we can look at the clusters each data point was assigned to
print(model.labels_)
```

```
plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1],c=model.labels_.astype(np.float))
plt.show()
```
